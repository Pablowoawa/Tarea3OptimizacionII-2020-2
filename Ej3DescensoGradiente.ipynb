```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
import math
# Define the original function
def f(x1, x2):
    return math.e**(x1+ 3*x2 - 0.1)+math.e**(x1 - 3*x2 - 0.1)+math.e**(- x1 - 0.1)

# Define gradient function (derivative function)
def gradient_x1(x1, x2):
    return math.e**(x1 - 3*x2 - 0.1) + math.e**(x1 + 3*x2 - 0.1) - math.e**(-x1 - 0.1)

def gradient_x2(x1, x2):
    return 3*math.e**(x1 + 3*x2 - 0.1) - 3*math.e**(x1 - 3*x2 - 0.1)
# Define learning rate
eta = 0.1
iterMax = 100
epsilon = 0.000001
# Define a list of tracks for x1, x2 and y.
x1_list = []
x2_list = []
y_list = []
# Define initial position
x1, x2 = -1,0
x1_list.append(x1)
x2_list.append(x2)
for i in range(1,iterMax+1):
    # Track list added to the corresponding track
    y_list.append(f(x1, x2))
    x1 = x1 - eta * gradient_x1(x1, x2)
    x2 = x2 - eta * gradient_x2(x1, x2)
    x1_list.append(x1)
    x2_list.append(x2)
    print(str(i)+" "+str(x1)+" "+str(x2)+" "+str(f(x1, x2)))
    if abs(x1_list[i] - x1_list[i - 1]) <= epsilon and abs(x2_list[i] - x2_list[i - 1]) <= epsilon:
        break
```
