```python
from mpl_toolkits.mplot3d import Axes3D
import numpy as np
import matplotlib.pyplot as plt
import sympy as sym
import math
# Define the original function
def f(x1, x2):
    return  100*(x2 - x1**2)**2 + ( 1 - x1)**2

# Define gradient function (derivative function)
def gradient_x1(x1, x2):
    return -400*x1*(x2 - x1**2) - 2*(1 - x1)

def gradient_x2(x1, x2):
    return 200*(x2 - x1**2)
# Define learning rate
eta = 0.001
# Define a list of tracks for x1, x2 and y.
x1_list = []
x2_list = []
y_list = []
# Define initial position
x1, x2 = -1.5,-1
itermax=200
for i in range(itermax):
    # Track list added to the corresponding track
    x1_list.append(x1)
    x2_list.append(x2)
    y_list.append(f(x1, x2))
    x1 = x1 - eta * gradient_x1(x1, x2)
    x2 = x2 - eta * gradient_x2(x1, x2)
    print(str(i)+" "+str(f(x1, x2)))
# display(x1_list)
# display(x2_list)
# display(y_list)

X1 = np.arange(-5, 5, 0.1)
X2 = np.arange(-5, 5, 0.1)
# grid.
# X1 looks like a column vector and expands along the line. The number of extended lines is the same as the number of X2 elements.
# X2 looks like a row vector and expands along the column. The number of columns expanded is the same as the number of X1 elements.
# Returns X1 and X2 after expansion. (After expansion, the shapes of X1 and X2 are the same.)
# meshgrid Extension Purpose: After expansion, we can get any combination of elements in X1 and X2. (eg Ptyhon zip)
X1, X2 = np.meshgrid(X1, X2)
Y = f(X1, X2)
fig = plt.figure()
# Create a 3D drawing object and draw on the parameter (figure).
ax = Axes3D(fig)
# Draw a surface map.
ax.plot_surface(X1, X2, Y, rstride=5, cstride=5, cmap="rainbow")
ax.plot(x1_list, x2_list, y_list, "bo--")
```
